```python
FXD_IMG =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(FXD_IMG)
```

```python
img_gray = cv2.imread('DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE) #gray scale so no channels present
```

```python
#resizing img
plt.imshow(FXD_IMG)
```

```python
FXD_IMG.shape
```

```python
new_img = cv2.resize(FXD_IMG,(1000,400))
plt.imshow(new_img)
```

```python
#resize by ratio
w_ratio = 0.8 # 80% of original and so on
h_ratio = 0.2
new_img = cv2.resize(FXD_IMG,(0,0), FXD_IMG,w_ratio,h_ratio)
plt.imshow(new_img)
```

```python
#flip images
new_img = cv2.flip(FXD_IMG,-1)
plt.imshow(new_img)
```
