```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_image():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='ABCDE',org=(50,300),fontFace=font,fontScale=5,color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return blank_img
```

```python
def display_image(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = load_image()
display_image(img)
```

```python
kernel = np.ones((5,5),dtype=np.uint8)  # 5x5 square kernel
result = cv2.erode(img, kernel, iterations=4) # 4 iteration of erosion
display_image(result)
```

```python
white_noise = np.random.randint(low=0, high=2, size=(600,600)) 
```

```python
# display_image(white_noise)
```

```python
white_noise = white_noise * 255
# display_image(white_noise)
```

```python
noise_img = white_noise + img
display_image(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel) # remove  background noise
```

```python
display_image(opening)
```

```python
black_noise = np.random.randint(low=0, high=2, size=(600,600))
black_noise = black_noise * -255
```

```python
black_noise_img = img + black_noise
black_noise_img[black_noise_img == -255] = 0
display_image(black_noise_img)
```

```python
closing = cv2.morphologyEx(black_noise_img, cv2.MORPH_CLOSE, kernel) # remove foreground noise
display_image(closing)
```
