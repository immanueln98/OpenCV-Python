## Contour Detection

```python
img = cv2.imread('DATA/internal_external.png',0)
plt.imshow(img , cmap='gray')
```

```python
image,coutours,hierarchy =cv2.findContours(img,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
external_contours = np.zeros(image.shape)

for i in range(len(coutours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(external_contours,coutours,i,255,-1)
    else:
        cv2.drawContours(external_contours,coutours,i,123,-1)

plt.imshow(external_contours,cmap='gray')
```

```python
internal_contours = np.zeros(image.shape)

for i in range(len(coutours)):  
    if hierarchy[0][i][3] != -1:
        cv2.drawContours(internal_contours,coutours,i,255,-1)

plt.imshow(internal_contours,cmap='gray')
```
